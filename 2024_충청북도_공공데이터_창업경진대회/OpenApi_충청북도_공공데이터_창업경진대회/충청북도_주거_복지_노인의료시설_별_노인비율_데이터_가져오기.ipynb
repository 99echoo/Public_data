```python
import requests
import csv

# CSV 파일 저장 경로 설정
csv_file_path= "C:/Users/user/Desktop/Public_data/2024_충청북도_공공데이터_창업경진대회/충청북도_주거_복지_노인의료시설_별_노인비율_데이터.csv"
# API 요청 기본 정보
url = 'http://apis.data.go.kr/6430000/ollrFcltyAcctoOdsnPopltnRateDataService/getollrFcltyAcctoOdsnPopltnRateData'
decoding = 'mW18J6e5gQ61JGt5YYgjiPhATkpOt/KJJxNjPUP3j/RoQF4VCrtduhvdMHs52rOOgdKh15FqXmnsPSsk8avWWA=='

# 칼럼명 동적 생성 함수
def fetch_initial_columns(json_data):
    first_item = json_data['body'][0]  # 첫 번째 아이템 접근
    return list(first_item.keys())  # 모든 키를 칼럼명으로 사용

# 최초 API 호출을 통해 칼럼명 설정
initial_params = {
    'serviceKey' : decoding, 
    'currentPage' : '1',
    'perPage' : '10',
}
response = requests.get(url, params=initial_params)
response.raise_for_status()
json_data = response.json()
columns = fetch_initial_columns(json_data)

# CSV 파일 오픈 및 헤더 작성
with open(csv_file_path, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(columns)  # 칼럼명을 헤더로 작성
    
    # 페이지별 데이터 수집 및 CSV 파일에 작성
    for page in range(1, 100):
        params = {
            'serviceKey' : decoding,
            'currentPage' : page,
            'perPage' : '10'
        }
        response = requests.get(url, params=params)
        response.raise_for_status()
        json_data = response.json()
        
        # 각 항목별 데이터를 CSV 파일에 추가
        for item in json_data['body']:
            row = [item.get(col, '') for col in columns]  # 동적 칼럼명을 사용하여 데이터 추출
            writer.writerow(row)

print(f"CSV 파일이 저장된 위치: {csv_file_path}")
```
