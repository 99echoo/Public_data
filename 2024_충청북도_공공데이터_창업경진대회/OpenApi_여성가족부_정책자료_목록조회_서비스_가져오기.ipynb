## 전체 데이터 로드

```python
import requests
import json
import csv

# CSV 파일 저장 경로 설정
csv_file_path = "C:/Users/user/Desktop/공모전/DataFrame.csv"

# API 요청 기본 정보
url = 'http://apis.data.go.kr/1383000/policy/subjectList'
params = {
    'serviceKey' : 'mW18J6e5gQ61JGt5YYgjiPhATkpOt/KJJxNjPUP3j/RoQF4VCrtduhvdMHs52rOOgdKh15FqXmnsPSsk8avWWA==',
    'type' : 'json',
    'numOFRows' : '10',
    'apiType' : 'equApi'
}

# CSV 파일 오픈 및 헤더 작성
with open(csv_file_path, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['담당자 이름', '부서명', '제목', '등록일', '전화번호', '조회수', 'URL'])
    
    # 1페이지부터 48페이지까지 반복
    for page in range(1, 49):  # 1부터 48까지 페이지 숫자
        params['pageNo'] = page
        response = requests.get(url, params=params)
        json_data = response.json()
        items = json_data['body'][0]['items']['item']
        
        # 각 항목별 데이터를 CSV 파일에 추가
        for item in items:
            writer.writerow([item.get('chrgrNm', ''), item.get('deptNm', ''), item.get('title', ''),
                             item.get('regDt', ''), item.get('telNo', ''), item.get('inqCnt', ''),
                             item.get('url', '')])

print(f"CSV 파일이 저장된 위치: {csv_file_path}")
```
